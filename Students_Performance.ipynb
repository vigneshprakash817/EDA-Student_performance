```python
g_cnt=df['Gender'].value_counts() # Gender count
```

```python
g_cnt
```

```python
total_students = df.shape[0]
```

```python
df['TotalStudents'] = total_students # total students
```

```python
df
```

```python
#Line chart

plt.plot(g_cnt.index,g_cnt.values)
plt.title('Gender wise total_students',fontdict={'fontsize':20,'fontname':'arial','fontweight':'bold','color':'blue'})
plt.xlabel('Gender')
plt.ylabel('total_students')
plt.grid(True)
plt.show()
```

```python
#Bar Chart
plt.bar(df['Ethnicity'],df['Absences'])
plt.title('Ethnicity wise Absences',fontdict={'fontsize':15,'fontname':'arial','fontweight':'bold','color':'Yellow'})
plt.xlabel('Ethnicity')
plt.ylabel('Absences')
plt.grid(True)
plt.show()
```

```python
#Histogram Chart
plt.hist(df['Age'],bins=3)
plt.title('Age wise Bin',fontdict={'fontsize':20,'fontname':'Cambria','fontweight':'bold','color':'Brown'})
plt.xlabel('Bin')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
#scatterplot chart
sns.scatterplot(x='ParentalSupport',y='GPA',data=data)
plt.title('ParentalSupport Wise GPA',fontdict={'fontsize':20,'fontname':'Cambria','fontweight':'bold','color':'Brown'})
```

```python
# Pie Chart
plt.pie(g_cnt,labels=g_cnt.values)
plt.title('Gender wise Count',fontdict={'fontsize':20,'fontname':'arial','fontweight':'bold','color':'blue'})
plt.grid(True)
plt.show()
plt.figure(figsize=(8,6))
```

```python
#Seaborn---Line Chart
sns.lineplot(x='Age',y='Absences',data=data)
plt.title('Age wise Absences',fontdict={'fontsize':16,'fontname':'Cambria','fontweight':'bold','color':'blue'})
```

```python
#Scatterplot chart
sns.scatterplot(x='Ethnicity',y='StudyTimeWeekly',data=data)
plt.title('Ethnicity Wise StudyTimeWeekly',fontdict={'fontsize':20,'fontname':'arial','fontweight':'bold','color':'blue'})
```

```python
#Bar chart
sns.barplot(x='Gender',y=g_cnt,data=data)
plt.title('Gender Wise Student count',fontdict={'fontsize':20,'fontname':'arial','fontweight':'bold','color':'Brown'})
```

```python
#Histogram chart
sns.histplot(data=data,x='GPA',bins=5,kde=True)
plt.title('GPA Wise Bin',fontdict={'fontsize':20,'fontname':'arial','fontweight':'bold','color':'red'})
```
