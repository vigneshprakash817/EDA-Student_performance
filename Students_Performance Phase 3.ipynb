SKEWNESS USING POWER TRANSFORMER FROM SKLEARN

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson')
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
df_transformed = pd.DataFrame(
    pt.fit_transform(df[numeric_cols]),
    columns=numeric_cols
)
```

```python
df_transformed.skew()
```

Plotting Histplot

```python
plt.figure(figsize=(20, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(1, len(numeric_cols), i + 1)
    sns.histplot(df_transformed[col], kde=True, bins=20, color='skyblue')
    plt.title(f"{col} (Transformed)")
    plt.xlabel('')
    plt.ylabel('')
```

```python
# Plot each column individually
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_transformed[col], kde=True, bins=20, color='teal')
    plt.title(f"{col} (Transformed)")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
X= df.drop('GPA',axis=1)
y= df['GPA']
```

```python
X
```

```python
y
```

```python
#Feature Selection
from sklearn.feature_selection import SelectKBest,f_regression
```

```python
skb= SelectKBest( score_func= f_regression, k='all')
```

```python
X_new=skb.fit_transform(X,y)
```

```python
selected_columns= X.columns[skb.get_support()]
selected_columns
```

```python
X_selected= X[selected_columns]
X_selected
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
X_scaled= ss.fit_transform(X_selected)
X_scaled
```

```python
#Training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X_scaled,y,test_size=0.2,random_state=40)
```

```python
print('X_train Shape:',X_train.shape)
print('X_test Shape:',X_test.shape)
print('Y_train Shape:',y_train.shape)
print('Y_test Shape:',y_test.shape)
```

```python
y.value_counts() 
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor()
```

```python
dtr.fit(X_train, y_train)
```

```python
y_pred_dtr= dtr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
MAE = mean_absolute_error(y_test,y_pred_dtr)
MSE = mean_squared_error(y_test,y_pred_dtr)
RMSE = np.sqrt(MSE)
R2= r2_score(y_test,y_pred_dtr)
```

```python
print("\nDecision Tree Regression Evaluation:")
print("MAE:", MAE)
print("MSE:", MSE)
print("RMSE:", RMSE)
print("R2 Score:", R2)
```

```python
#Support Vector Regressor
from sklearn.svm import SVR
```

```python
svr=SVR()
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred_svr= svr.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,y_pred_svr)
MSE = mean_squared_error(y_test,y_pred_svr)
RMSE = np.sqrt(MSE)
R2= r2_score(y_test,y_pred_svr)
```

```python
print("\nSupport Vector Regression Evaluation:")
print("MAE:", MAE)
print("MSE:", MSE)
print("RMSE:", RMSE)
print("R2 Score:", R2)
```

```python
#Ensemble Learning ---> RandomForestregressor
from sklearn.ensemble import RandomForestRegressor
```

```python
rfg=RandomForestRegressor()
```

```python
rfg.fit(X_train,y_train)
```

```python
y_pred_rfg= rfg.predict(X_test)
```

```python
MAE = mean_absolute_error(y_test,y_pred_rfg)
MSE = mean_squared_error(y_test,y_pred_rfg)
RMSE = np.sqrt(MSE)
R2= r2_score(y_test,y_pred_rfg)
```

```python
print("\nRandom forest Regression Evaluation:")
print("MAE:", MAE)
print("MSE:", MSE)
print("RMSE:", RMSE)
print("R2 Score:", R2)
```
